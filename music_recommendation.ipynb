```python
import numpy
import pandas as pd
```

```python
df= pd.read_csv('ex.csv')
```

```python
df.head(5)
```

```python
df.dropna(inplace=True)
#removing missing values
```

```python
df= df.drop_duplicates()
```

```python
df.head(5)
df.duplicated().sum()
```

```python
df.shape
#new dimensions of the dataframe after dropping duplicate values
```

```python
l=[]
for i in df['User-Rating']:
    l.append(i[1:3])
l
```

```python
df['Album/Movie']= df['Album/Movie'].str.replace(' ','')
df['Singer/Artists']= df['Singer/Artists'].str.replace(' ','')
df
```

```python
df['Singer/Artists']= df['Singer/Artists'].str.replace(' ','')
df
```

```python
df['tags']= df['Singer/Artists']+' '+df['Genre']+' '+df['Album/Movie']+' '+df['User-Rating']
df['tags'][0]
```

```python
new_df=df[['Song-Name','tags']]
new_df
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:x.lower())
#applying a lambda function to convert all the values in the 'tags' column to lowercase.
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000)
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key= lambda x:x[1])
#in decreasing order to keep thew highest similarity value on the top
```

```python
new_df.rename(columns={'Song-Name':'title'},inplace= True)
```

```python
def recommend(music):
    music_index= new_df[new_df['title']==music].index[0]
    distances= similarity[music_index]
    music_list= sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title)

```

```python
recommend('Tere Bin')
```

```python
recommend('Do Peg Maar')
```
